```python
# concate and merge
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                     'value': [5, 6, 7, 8]})
df1
```

```python
df2
```

```python
df1.merge(df2,left_on='lkey',right_on='rkey',how='inner')
```

```python
df1.merge(df2,left_on='lkey',right_on='rkey',how='right')
```
