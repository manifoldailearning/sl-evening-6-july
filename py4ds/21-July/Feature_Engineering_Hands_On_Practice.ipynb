```python
df[df['PLAYING ROLE']=='Batsman'].groupby('TEAM')["SOLD PRICE"].mean().sort_values(ascending=False)
```

```python
# Which player had the highest difference between base price and sold price ?
df['Difference'] = df['SOLD PRICE'] - df['BASE PRICE']
df.sort_values('Difference', ascending=False).head(1)
```

```python
df['diff'] = df['SOLD PRICE'] - df['BASE PRICE']
df[df['diff'] == df['diff'].max()][['PLAYER NAME', 'BASE PRICE', 'SOLD PRICE',"diff"]] # boolean masking
```

```python
# Calculate the difference between soldprice and baseprice
df['price_difference'] = df['SOLD PRICE'] - df['BASE PRICE']
 # Identify the player with the highest difference
max_diff_player = df.iloc[df['price_difference'].idxmax()]
```
