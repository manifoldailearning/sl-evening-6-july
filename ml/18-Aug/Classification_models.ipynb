```python
# Logistic Regression

import numpy as np
import matplotlib.pyplot as plt


def linear_model(X, W, b):
    return np.dot(X, W) + b

def logistic(z):
    return 1 / (1 + np.exp(-z))
```

```python
W = 5
b = 2

x = np.linspace(-10,10,100)
z= linear_model(x,W,b)
x
```

```python
plt.plot(x,z)
```

```python
output = logistic(z)
plt.plot(x,output)
plt.show()
```

# KNN

- Manhattan : L1 norm
   - Dimension is higher
   - Features of different scales
   - Sparse data

- Euclidean Distance: L2 norm
  - Low dimensional data
  - features on similar scale
  - Dense data

- Minkowski Distance : Lp norm
  - Experimentation


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3, p=2)
knn.fit(X_train,y_train)

y_pred_test = knn.predict(X_test)

f1_score(y_test,y_pred_test)
```

```python
params = {"n_neighbors": [3,4,5,6,7,8],
          "p":[1,2,3]}

grid = GridSearchCV(KNeighborsClassifier(),param_grid=params,cv=5,verbose=2,scoring="f1")
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
best_model = grid.best_estimator_
y_pred_test = best_model.predict(X_test)
f1_score(y_test,y_pred_test)
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
pipeline = Pipeline(steps=[("scaler",StandardScaler()),
                           ("knn", KNeighborsClassifier())])

params = {"knn__n_neighbors": [3,4,5,6,7,8],
          "knn__p":[1,2,3]}
grid = GridSearchCV(pipeline,param_grid=params,cv=5,verbose=2,scoring="f1")
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {'knn__n_neighbors': np.arange(1, 21)}

# Create a pipeline with a scaler and KNN classifier
pipeline = Pipeline([
    ('scaler', StandardScaler()),    # Standardize the data
    ('knn', KNeighborsClassifier())  # KNN classifier
])

# Initialize GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid, cv=10, scoring='accuracy')

# Fit GridSearchCV to the data
grid_search.fit(X_train, y_train)

# Get the best k value
optimal_k = grid_search.best_params_['knn__n_neighbors']
print(f"The optimal number of neighbors is {optimal_k}")

# Get the best score
best_score = grid_search.best_score_
print(f"The best cross-validated accuracy is {best_score}")

# Get the cross-validation results
cv_results = grid_search.cv_results_

# Plot the cross-validation accuracy for different k values
plt.figure(figsize=(10, 6))
plt.plot(param_grid['knn__n_neighbors'], cv_results['mean_test_score'], marker='o')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Cross-Validation Accuracy')
plt.title('KNN Cross-Validation Accuracy for Different K Values')
plt.show()
```

# Voting Classifier

```python
from sklearn.ensemble import VotingClassifier

knn = KNeighborsClassifier(n_neighbors=3, p=2)
tree = DecisionTreeClassifier(class_weight={0:0.36,1:0.62})
svm = SVC(kernel="linear")

vote = VotingClassifier(estimators=[("knn",knn),("tree",tree),("svm",svm)], voting="hard")
vote.fit(X_train,y_train)
```

```python
y_pred_test =vote.predict(X_test)
f1_score(y_test,y_pred_test)
```

```python
from sklearn.ensemble import VotingClassifier

knn = KNeighborsClassifier(n_neighbors=3, p=2)
tree = DecisionTreeClassifier(class_weight={0:0.36,1:0.62})
svm = SVC(kernel="linear", probability=True)

vote = VotingClassifier(estimators=[("knn",knn),("tree",tree),("svm",svm)], voting="soft")
vote.fit(X_train,y_train)
```

```python
y_pred_test =vote.predict(X_test)
f1_score(y_test,y_pred_test)
```

# Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier
logistic = LogisticRegression()

bag = BaggingClassifier(estimator=logistic, n_estimators=100, max_features=0.8, max_samples=0.8)
bag.fit(X_train,y_train)

y_pred_test = bag.predict(X_test)
f1_score(y_test,y_pred_test)
```

```python
# BaggingClassifier + estimator of DecisionTreeClassifier = RandomForestClassifer
bag = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, max_features=0.8, max_samples=0.8)
bag.fit(X_train,y_train)

y_pred_test = bag.predict(X_test)
f1_score(y_test,y_pred_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100, max_features=0.8, max_samples=0.8)
forest.fit(X_train,y_train)

y_pred_test = forest.predict(X_test)
f1_score(y_test,y_pred_test)
```
