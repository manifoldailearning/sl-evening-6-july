```python
# Logistic Regression

import numpy as np
import matplotlib.pyplot as plt


def linear_model(X, W, b):
    return np.dot(X, W) + b

def logistic(z):
    return 1 / (1 + np.exp(-z))
```

```python
W = 5
b = 2

x = np.linspace(-10,10,100)
z= linear_model(x,W,b)
x
```

```python
plt.plot(x,z)
```

```python
output = logistic(z)
plt.plot(x,output)
plt.show()
```

```python
# Pipeline
# features - StandardScaler, fit on model

# multiple preprocessing steps, and finally add the algorithm
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps=[("scaler", StandardScaler()),
                       ("logistic", LogisticRegression())])

# fit --> on all the preprocessing step - it will perform fit_transform
# model - fit

# Prediction--> transform on all the preprocessing, and predict on model
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()

# no mandate rule of scaling
tree.fit(X_train,y_train)
```

```python
#plot the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(30,30))
plot_tree(tree)
plt.show()
```

```python
tree.get_depth()
```

```python
y_pred_train = tree.predict(X_train)
accuracy_score(y_train,y_pred_train)
```

```python
y_pred_test = tree.predict(X_test)
accuracy_score(y_test,y_pred_test)
```

```python
# Adjustment of parameters - Hyperparameters
# Hyperparameters are the parameters of ML model which we have to initialize and will not be
# determined as a result of training

tree = DecisionTreeClassifier(criterion= "gini",max_depth=4) # gini
# no mandate rule of scaling
tree.fit(X_train,y_train)
y_pred_train = tree.predict(X_train)
y_pred_test = tree.predict(X_test)

print(f"accuracy score of training data is {accuracy_score(y_train,y_pred_train)}")
print(f"accuracy score of testing data is {accuracy_score(y_test,y_pred_test)}")

plt.figure(figsize=(30,30))
plot_tree(tree)
plt.show()
```

```python
# Adjustment of parameters - Hyperparameters
# Hyperparameters are the parameters of ML model which we have to initialize and will not be
# determined as a result of training

tree = DecisionTreeClassifier(criterion="entropy",max_depth=3)
# no mandate rule of scaling
tree.fit(X_train,y_train)
y_pred_train = tree.predict(X_train)
y_pred_test = tree.predict(X_test)
print(f"accuracy score of training data is {accuracy_score(y_train,y_pred_train)}")
print(f"accuracy score of testing data is {accuracy_score(y_test,y_pred_test)}")

plt.figure(figsize=(30,30))
plot_tree(tree)
plt.show()
```

```python
# How to find the best hyperparameters ?

# GridSearchCV
from sklearn.model_selection import GridSearchCV

params = {"max_depth":[3,4,5,6],
          "criterion":["gini","entropy","log_loss"],
          "min_samples_leaf":[2,3,4]}
```

```python
gridcv = GridSearchCV(DecisionTreeClassifier(),param_grid=params,cv=5,verbose=2, scoring="f1_score")
gridcv.fit(X_train,y_train)
```

```python
gridcv.best_params_
```

```python
gridcv.best_score_
```

```python
# RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
random_cv = RandomizedSearchCV(estimator = DecisionTreeClassifier(),
                               param_distributions = params,
                               cv = 5, n_iter = 10,
                               scoring = 'accuracy',verbose=2)
random_cv.fit(X_train,y_train)
```
