```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# have the features in a 1d format
X_train_new = X_train.reshape(60000,28*28) # 60000,784
X_test_new = X_test.reshape(10000,28*28)
```

```python
X_train_new.shape, X_test_new.shape, y_train.shape, y_test.shape
```

```python
# Preprocessing of the target
# Create One Hot encoding for target data
from tensorflow.keras.utils import to_categorical
y_train_transform = to_categorical(y_train, num_classes=10)
y_test_transform = to_categorical(y_test, num_classes=10)
```

```python
y_train_transform[:5]
```

```python
np.max(X_train_new),np.min(X_train_new)
```

```python
# Input data preprocessing
X_train_new = X_train_new/255
X_test_new = X_test_new/255
```

```python
model = Sequential()
model.add(Dense(units = 128, activation='sigmoid', input_shape=[784,]))
model.add(Dense(units = 10, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer= tf.keras.optimizers.SGD(learning_rate=0.1),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train_new, y_train_transform, epochs=10,
          validation_data=(X_test_new, y_test_transform))
```

```python
model.evaluate(X_test_new, y_test_transform)
```

```python
h = history.history
```

```python
def Generate_Data(w, b, instances):
    """Generate y = Xw + b + noise."""
    X = tf.zeros((instances, w.shape[0]))
    X += tf.random.normal(shape=X.shape)
    y = tf.matmul(X, tf.reshape(w, (-1, 1))) + b
    y += tf.random.normal(shape=y.shape, stddev=0.01) # Noise
    y = tf.reshape(y, (-1, 1)) # Flattening
    return X, y

w_true = tf.constant([5.0,3.0]) # w1, w2
b_true = 4.0
features, labels = Generate_Data(w_true, b_true, 100)
# (100,2) (100)
```

```python
features.shape
```

```python
features[:2]
```

```python
labels[:2]
```

```python
w = tf.Variable(tf.random.uniform([2,1]), trainable=True)
b = tf.Variable(tf.random.uniform([1]), trainable=True)
w,b
```

```python
# Forward pass
def model(X): # Linear Summation
  return tf.tensordot(X,w,axes=1) + b

# Loss function
def loss_fn(y_true, y_pred):
  return tf.reduce_mean(tf.square(y_true - y_pred))
```

```python
lr = 0.01
n = 10
loss = loss_fn

for epoch in range(n):
  for X,y in zip(features,labels):
    with tf.GradientTape() as tape:
      y_pred = model(X)
      loss_value = loss(y, y_pred)

    dw,db = tape.gradient(loss_value,[w,b])
    w.assign_sub(lr*dw) # w = w - lr*dw
    b.assign_sub(lr*db) # b = b = lr*db
  loss_value = loss(labels, model(features))
  print(f'Epoch: {epoch+1}, Loss: {loss_value.numpy()}')
```

# Regularizer & Dropout

```python
from tensorflow.keras.regularizers import L1, L2
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(units = 128, activation='sigmoid', input_shape=[784,]))
model.add(Dense(units = 128, activation='sigmoid', kernel_regularizer = L1(l1=0.01)))
model.add(Dense(units = 10, activation='softmax'))
model.summary()
```

# Dropout

```python
model = Sequential()
model.add(Dense(units = 128, activation='sigmoid', input_shape=[784,]))
model.add(Dense(units = 128, activation='sigmoid', kernel_regularizer = L1(l1=0.01)))
model.add(Dropout(0.2))
model.add(Dense(units = 10, activation='softmax'))
model.summary()
```
